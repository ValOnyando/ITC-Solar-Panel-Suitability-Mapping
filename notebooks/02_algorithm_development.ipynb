## 2. Solar Energy Calculations

### Mathematical Formula

$$E = A \times H \times \eta \times (1 - S)$$

Where:
- $E$ = Energy production (kWh)
- $A$ = Roof area (m²)
- $H$ = Solar irradiance (kWh/m²/year)
- $\eta$ = Panel efficiency (%)
- $S$ = Shading factor (0-1)
